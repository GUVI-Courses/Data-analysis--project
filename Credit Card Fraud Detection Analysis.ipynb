# Data Preprocessing

## Data Cleaning

```python
df.isnull().sum()
```

```python
null_count = df.isna().sum()
null_cols = null_count[null_count>0].index.to_list()
print(null_cols)
```

### Imputation - fill the missing values

```python
null_cat_cols = [col for col in null_cols if col in categorical_columns]
print(null_cat_cols)
```

```python
# Impute the categorical columns with MODE
for col_name in null_cat_cols:
    df[col_name].fillna(df[col_name].mode()[0], inplace=True)
```

```python
# Impute 'Cardholder_Age', 'Cardholder_Monthly_Income' numerical columns with MEAN
df.fillna({col: df[col].mean() for col in ['Cardholder_Age', 'Cardholder_Monthly_Income']}, inplace=True)
```

```python
# Impute 'Cardholder_Average_Spend' with MEDIAN
df['Cardholder_Average_Spend'].fillna(df['Cardholder_Average_Spend'].median(), inplace=True)
```

```python
df.head()
```

```python
df[['Cardholder_Age', 'Credit_Limit']] = df[['Cardholder_Age', 'Credit_Limit']].astype(int)
```

```python
df.dtypes
```

### Data pruning

```python
# Drop rows with misisng 'Transaction_Amount'
df.dropna(subset=['Transaction_Amount'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

### Check outliers

```python
sns.boxplot(x=df['Transaction_Amount'])
```

```python
sns.boxplot(x=df['Cardholder_Average_Spend'])
```

### Remove outliers

```python
Q1 = df['Cardholder_Average_Spend'].quantile(0.25)
Q3 = df['Cardholder_Average_Spend'].quantile(0.75)

IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

print(lower_bound, upper_bound)
```

```python
df['Cardholder_Average_Spend'].max()
```

```python
df = df[(df['Cardholder_Average_Spend'] >= lower_bound) & (df['Cardholder_Average_Spend'] <= upper_bound)]
df['Cardholder_Average_Spend'].max()
```

```python
sns.boxplot(x=df['Cardholder_Average_Spend'])
```

### Square root transformation to treat the skewed data

```python
skewed_cols = ['Transaction_Amount', 'Cardholder_Monthly_Income', 'Cardholder_Average_Spend']
```

```python
df[skewed_cols].plot(kind='kde', subplots=True)
```

```python
for col in skewed_cols:
    df[col] = (df[col])**(1/2)  # square root
df[skewed_cols] = df[skewed_cols].round(2)
```

## Feature Selection

```python
df.columns
```

```python
# Select features and target
X = df.drop(['Transaction_ID', 'Transaction_DateTime', 'Is_Fraudulent'], axis=1)
y = df['Is_Fraudulent']
```

```python
X.columns
```

```python
X.head()
```

## Feature Engineering

```python
x_obj_cols = [col for col in categorical_columns if col!='Is_Fraudulent']
print(x_obj_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

X[x_obj_cols] = X[x_obj_cols].apply(LabelEncoder().fit_transform)
X.head()
```

```python
X.dtypes
```

## Split data into training and testing

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
```

```python
y.value_counts()
```

```python
x_train.head(3)
```

```python
x_test.head(3)
```

## Feature Scaling

```python
print(numerical_columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train[numerical_columns] = scaler.fit_transform(x_train[numerical_columns]).round(2)
x_test[numerical_columns] = scaler.transform(x_test[numerical_columns]).round(2)
```

```python
x_train.sample(5)
```

# Model Building

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score

model = SVC()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(f"Accuracy = {accuracy_score(y_test, y_pred)}")
```

```python
model = SVC(class_weight='balanced')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(f"Recall = {recall_score(y_test, y_pred, pos_label='Yes')}")
```
